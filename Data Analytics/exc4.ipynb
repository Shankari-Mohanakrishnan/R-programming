```r

data <- read.csv("bhp.csv", header =TRUE)
```

```r
data
```

```r
data$class<- ifelse(data$CLASS == 1, "Good", "Failure")
```

```r
data
```

```r
set.seed(7457)
ind = sample(2, nrow(data), replace = TRUE, prob=c(0.7, 0.3))
train = data[ind == 1, ] 
test = data[ind == 2, ] 

```

```r
str(train)
str(test)
```

```r
logit<-glm(CLASS~., family=binomial,data=train)

```

```r
summary(logit)
```

```r
# deviance  - The deviance result gives the following information:
# q1 = -2.409e-06
# q3= 2.409e-06 
# median = 2.409e-06 
# min = -2.409e-06
# max = 2.409e-06
coef(logit)
```

```r
train.prob<-predict(logit, type="response")
train.prob
```

```r
train.prob[1:6]
```

```r
# Classification of cut off can be done by when a value is equal to or greater than 0.5 they would be value =  1 
# else if value is less than 0.5 then value= 0
# In the above example 3, 4, 5, 6, 12 will belong to value=1 and 7 will belong to value=0
```
