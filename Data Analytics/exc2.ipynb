```r
data<-read.csv("gasmf.csv", header =TRUE)
data
x1 <-data[,2]
x2 <-data[,3]
x3 <-data[,4]
x4 <-data[,5]
x5 <-data[,6]
x6 <-data[,7]
x7 <-data[,8]
x8 <-data[,9]
x9 <-data[,10]
x10 <-data[,11]
x11 <-data[,12]
y<-data[,13]
```

```r
m1 <-lm(y~x1+x6)
summary(m1)
# Regression equation : 32.88451-0.053148x1+0.959223x6
```

```r
var<-anova(m1)
summary(var)
```

```r
# R-square= 0.7873
# Adj-R= 0.7726
```

```r
confint(m1,"x1", level = 0.95)
```

```r
# According to t-test in the model both x1 and x6 <0.05. So both the variables are significant.
t.test(x1, x6, paired= FALSE, mu =0)
```

```r
# Regression equation : 32.88451-0.053148(275)+0.959223(2)
# =20.186846
```

```r
predict(m1, data.frame(x1=275,x6 = 2), interval="confidence",level= 0.95)
```
